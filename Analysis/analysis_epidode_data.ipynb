```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
```

```python
episode_data = []
with open('result_clusgym_ver22_expt1/episode_data.txt') as fh:
    for line in fh:
        line = line.strip()
        episode_data.append(line)
print("Total lines:", len(episode_data))

final_episode_list = []
for i in range(len(episode_data)):

    episode_data_list = episode_data[i].split(':') #split the left square bracket [


    episode_data_list_new = []
    for i, item in enumerate(episode_data_list):
        if i == 0:
            episode_data_list_new.append(item)
        else:
            pattern = r':,'
            item_split = re.split(pattern, item)
            episode_data_list_new.append(item_split)

    episode_data_list_flatten = []
    for i, item in enumerate(episode_data_list_new):
        if i == 0:
            episode_data_list_flatten.append(item)
        else:
            split_item =  item[0].split(",")
            for j in split_item:
                episode_data_list_flatten.append(j)

    episode_list_float_int = []
    for i, item in enumerate(episode_data_list_flatten):
        if i % 2 == 1 :
            if i in [3, 21, 23, 27]:
                item  =  float(item)
            else:
                item = int(item)
            episode_list_float_int.append(item)
        else:
            episode_list_float_int.append(item)

    #print(episode_list_float_int)
    final_episode_list.append(episode_list_float_int)
print(len(final_episode_list))
print(final_episode_list[0])
print(final_episode_list[-1])
        
```
